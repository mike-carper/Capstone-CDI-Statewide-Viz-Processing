```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd
import datetime as dt
```

```python
fileName = 'C:/Users/mcarp/OneDrive/Documents/tracts.csv'
tracts = pd.read_csv(fileName)
dftracts=pd.DataFrame(tracts)
```

```python
dftracts.head()
```

```python
def plot_indicators(tractID):
    df = dftracts[dftracts['TractID']==tractID]
    for x in df.columns[4:]:
        plt.plot(df.Year,df['{}'.format(x)],'o-')
        plt.title('Tract {}: {}'.format(tractID,x))
        plt.xlabel('Year')
        plt.ylabel('{}'.format(x))
        plt.xticks(range(2010,2020))
        plt.show()
```

```python
tractIDList=[302.0,303.01,303.02,304.01,304.02,305.02,305.04,305.05,316.0,320.0,201.04,201.05,201.06,201.07,221.0,222.0,223.0]
for tractID in tractIDList:
    plot_indicators(tractID)
```
