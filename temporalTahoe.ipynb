```python
import seaborn as sns
```

```python
fig = plt.figure()
fig.set_size_inches(20, 28)
i=1
fig.subplots_adjust(hspace=0.01, wspace=0.01)
for n in [302.0,303.01,303.02,304.01,304.02,305.02,305.04,305.05,316.0,320.0]:
    df=dftracts[dftracts['TractID']==n]
    ax = fig.add_subplot(6, 3, i, title='Affordable Rent: El Dorado Tract {}'.format(n))
    sns.lineplot(x=df['Year'], y=df['Affordable Rent'],ax=ax)
    i+=1
for t in [201.04,201.05,201.06,201.07,221.0,222.0,223.0]:
    df=dftracts[dftracts['TractID']==t]
    ax = fig.add_subplot(6, 3, i, title='Affordable Rent: Placer Tract {}'.format(t))
    sns.lineplot(x=df['Year'], y=df['Affordable Rent'],ax=ax)
    i+=1
fig.tight_layout()
plt.show()
```

```python
def plot_indicators(tractID):
    df = dftracts[dftracts['TractID']==tractID]
    for x in df.columns[4:]:
        plt.plot(df.Year,df['{}'.format(x)],'o-')
        plt.title('Tract {}: {}'.format(tractID,x))
        plt.xlabel('Year')
        plt.ylabel('{}'.format(x))
        plt.xticks(range(2010,2020))
        plt.show()
```
