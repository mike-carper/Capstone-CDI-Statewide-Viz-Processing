```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd
import datetime as dt
```

```python
fileName = 'C:/Users/mcarp/OneDrive/Documents/tracts.csv'
tracts = pd.read_csv(fileName)
dftracts=pd.DataFrame(tracts)
```

```python
def plot_indicators(tract):
    df = dftracts[dftracts['Tract']==tract]
    for x in df.columns[4:]:
        plt.plot(df.Year,df['{}'.format(x)],'o-')
        plt.title('Tract {}: {}'.format(df.TractID[0],x))
        plt.xlabel('Year')
        plt.ylabel('{}'.format(x))
        plt.xticks(range(2010,2020))
        plt.show()
```

```python
tractList=[6017030200,6017030301,6017030302, 6017030401,6017030402,6017030502,6017030504,6017030505,6017031600,6017032000,6061020104,6061020105,6061020106,6061020107,6061022100,6061022200,6061022300]
for tract in tractList:
    plot_indicators(tract)
```
